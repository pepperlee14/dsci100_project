**Predicting Heart Disease Risk through Resting Blood Pressure, Serum Cholestoral Level, Maximum Heart Rate Achieved, and ST Depression Induced by Exercise Relative to Rest.**

**INTRODUCTION**

Coronary heart disease (CHD) is one of the most prevalent conditions, affecting about 8% of the global population (Søjbjerg, et al., 2024). Various symptoms that cause significant emotional distress are often accompanied by CHD, leading to an increased rate of psychological issues such as anxiety and depression. As CHS progresses, physiological symptoms such as dyspnea or chest pain arise, which negatively affects individuals’ ability to perform daily and social life (Karataş, 2017). Therefore, this study aims to predict the presence of heart disease from different health aspects. The research question addressed is: Can the presence of heart disease be accurately predicted using resting blood pressure, serum cholesterol levels, maximum heart rate achieved during exercise, and ST depression induced by exercise relative to rest? A dataset of 425 anonymized medical records of patients in Cleveland, Ohio is used, sourced from University of California, Irvine. The dataset consists of originally fourteen variables. Classification with K-nearest neighbors (KNN) algorithm will be performed, using a new dataset consists of five variables of num (diagnosis of heart disease), trestbps (resting blood pressure), chol (serum cholestoral level), thalach (maximum heart rate achieved), and oldpeak (ST depression induced by exercise relative to rest). Results will be visualized with scatterplot and confusion matrix.

**PRELIMINARY EXPLORATORY DATA ANALYSIS**

*STEP 1: Loading packages of `tidyverse` and `tidymodels` into R in order to perform analysis with useful functions.*

```r
library(tidyverse)
library(tidymodels)
```

*STEP 2: Reading files into R*

```r
set.seed(100)

clevelandhd <- read_csv("data/heart_disease/processed.cleveland.data", col_names = FALSE)
clevelandhd
```

*STEP 3: Naming all columns and select the variables needed, creating a new dataset.*

```r
set.seed(100)

clevelandhd <- clevelandhd|>
    rename(age = X1,
           sex = X2,
           cp = X3,
           trestbps = X4,
           chol = X5,
           fbs = X6,
           restecg = X7,
           thalach = X8,
           exang = X9,
           oldpeak = X10,
           slope = X11,
           ca = X12,
           thal = X13,
           num = X14) |>
    select(trestbps, chol, thalach, oldpeak, num) |>
    mutate(num = as.factor(num)) |>
    mutate(num = fct_recode(num, "Yes" = "0", "No" = "1", "No" = "2", "No" = "3", "No" = "4"))
clevelandhd
```

*STEP 4: Spliting 75% of the original data into training data and 25% into testing data. Number of rows in each dataset and proportions of results in the training dataset are checked.*

```r
set.seed(100)

clevelandhd_split <- initial_split(clevelandhd, prop = 0.75, strata = num)
clevelandhd_training <- training(clevelandhd_split)
clevelandhd_testing <- testing(clevelandhd_split)
```

```r
glimpse(clevelandhd_training)
```

```r
glimpse(clevelandhd_testing)
```

```r
hd_proportions <- clevelandhd_training |>
    group_by(num) |>
    summarize(n = n()) |>
    mutate(percent = 100*n/nrow(clevelandhd_training))
hd_proportions
```

```r
clevelandhd_training
```

*STEP 5: Finding the most suitable and accurate K number to perform classification for the above dataset using the method of visualization.*

```r
set.seed(100)

clevelandhd_recipe <- recipe(num ~ trestbps + chol + thalach + oldpeak, data = clevelandhd_training) |>
    step_scale(all_predictors()) |>
    step_center(all_predictors())

knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
    set_engine("kknn") |>
    set_mode("classification")

clevelandhd_vfold <- vfold_cv(clevelandhd_training, v = 5, strata = num)

k_vals <- tibble(neighbors = seq(from = 1, to = 150))

clevelandhd_tuning_result <- workflow() |>
    add_recipe(clevelandhd_recipe) |>
    add_model(knn_tune) |>
    tune_grid(resamples = clevelandhd_vfold, grid = k_vals) |>
    collect_metrics()

accuracies <- clevelandhd_tuning_result |>
    filter(.metric == "accuracy")

accuracy_vs_k <- ggplot(accuracies, aes(x = neighbors, y = mean)) +
                             geom_point() +
                             geom_line() +
                             labs(x = "Neighbors", y = "Accuracy Estimate")
accuracy_vs_k
```

```r
accuracy_vs_k_plot <- ggplot(accuracies, aes(x = neighbors, y = mean)) +
                             geom_point() +
                             geom_line() +
                             labs(x = "Neighbors", y = "Accuracy Estimate") +
                            xlim(25,50)
accuracy_vs_k_plot                          
```

*Visualiaztion result:*

The estimated accuracy of the above model is the highest (around 76%) when K = 46. 

**METHODS**

In this study, classification with K-nearest neighbors (KNN) algorithm will be performed. The presence of heart disease, which is the categorical variable of interest, will be predicted, based on multiple variables of resting blood pressure, serum cholesterol levels, maximum heart rate achieved during exercise, and ST depression induced by exercise relative to rest. For this analysis, K = 46 will be applied as the number of neighbors. To visualize the results of KNN classification, scatterplots and confusion matrix will be used to show the relationship between the training and testing data, including model accuracies and data distributions.

**EXPECTED OUTCOME AND SIGNIFICANCE**

The accuracy of predicting the presence of heart disease using the variables of resting blood pressure, serum cholesterol levels, maximum heart rate achieved during exercise, and ST depression induced by exercise relative to rest is aimed to find in this study. This study aim for above 70% of accuracy. If successful, this predictive model can be a non-invasive reference of early detection of heart disease. The model can also be applied to other population and settings, and provides a way for future researches to include additional variables that might enhance the accuracy of CHD predictions.
